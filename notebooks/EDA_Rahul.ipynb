```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import set_style
set_style("whitegrid")
from scipy.stats import norm
```

Define function for calculating premium price for a call option using Black Scholes Model

```python
## Where:
## S = the current stock price
## K = the strike price of the option
## r = the risk-free-rate
## T = time to maturity, or time to expiration
## sigma = annualized volatility of the asset's returns
## N = A normal distribution

def BS_CALL(S,K,T,r,sigma):
    N = norm.cdf
    d1 = (np.log(S/K) + (r + sigma**2/2)*T)/(sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    return S*N(d1) - K*np.exp(-r*T)*N(d2)
```

Define function for calculating premium price for a put option using Black Scholes Model

```python
def BS_PUT(S,K,T,r,sigma):
    N = norm.cdf
    d1 = (np.log(S/K) + (r + sigma**2/2)*T)/(sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    return K*np.exp(-r*T)*N(-d2) - S*N(-d1)
```
